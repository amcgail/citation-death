# let's find a few interesting examples

Maybe just a random sample of 5 of the top cited docs in the 1970s?

```python
from knowknow import *
```

```python
db = Dataset('sociology-wos-74b')
```

```python
db.groupings.append((
    'fy',
    {
        k: (k//10)*10
        for k in range(1900,2020)
    }
))
```

```python
things = sorted( db.items('c'), key=lambda x: -db(c=x, fy=1970).docs )[:18]
```

```python
sorted(things)
```

```python
things = ['Becker, H.|outsiders studies so',
 'Berger, P.|social construction',
 'Blalock, H.|social statistics',
 'Blau, P.|am occupational stru',
 'Blau, P.|exchange power socia',
 'Coleman, J.|equality ed opportun',
 'Durkheim, E.|division labor soc',
 'Goffman, E.|asylums',
 'Homans, G.|social behavior its',
 'Kuhn, T.|structure sci revolu',
 'Lenski, G.|religious factor',
 'Lipset, S.|political man',
 'Merton, R.|social theory social',
 'Parsons, T.|social system',
 'Parsons, T.|structure social act',
 'Weber, M.|ec soc',
 'Weber, M.|protestant ethic spi',
 'Weber, M.|theory social ec org']
```

```python
def makeit(x):
    n, m = x.split("|")
    return "%s %s" % (n, m.title())
```

```python
viz.yearly_counts?
```

```python
db = Dataset('sociology-wos-74b')
```

```python
viz.yearly_counts(things, db, 'top9.1970s', cols=3, print_names=makeit)
```

# what if we look at these as a proportion of all docs?

```python
a = db.trend('c', 'Berger, P.|social construction').docs(1960,2015) / np.array([ db(fy=YY).docs for YY in range(1960,2016) ])
```

```python
np.max(a)
```

```python
plt.plot(a)
```

```python
max( [
    db(fy=YY,c='Berger, P.|social construction').cits / db(fy=YY).cits
    for YY in range(1960,2020)
] )
```

```python
"%0.1f%%" % (a*1000)
```

```python
viz.yearly_counts(things, db, 'top9.1970s.p', cols=3, print_names=makeit, yearly_prop=True, count_unit='docs')
```
