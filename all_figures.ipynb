```python
def pretty_time_delta(seconds):
    sign_string = '-' if seconds < 0 else ''
    seconds = abs(int(seconds))
    days, seconds = divmod(seconds, 86400)
    hours, seconds = divmod(seconds, 3600)
    minutes, seconds = divmod(seconds, 60)
    if days > 0:
        return '%d days ago' % (days)
    elif hours > 0:
        return '%d hours ago' % (hours)
    elif minutes > 0:
        return '%d minutes ago' % (minutes)
    else:
        return '%d seconds ago' % (seconds)
```

```python
import os
import time
from datetime import datetime

md = "# Table of Contents\n"

for uid in sorted(DOCS):
    d = DOCS[uid]
    if 'fn' not in d:
        continue
        
    files = list(Path(BASEDIR, 'figures').glob("%s.png" % d['fn']))
    
    name = d['name']
    if not len(files):
        name += " (no files)"
    else:
        mtime = min( os.path.getmtime(str(f)) for f in files )
        name += " (%s)" %pretty_time_delta(
            (datetime.now() - datetime.fromtimestamp(mtime)).total_seconds()
        )
    
    
    md += "* [%s](#%s)\n" % (name, uid)
    
display(Markdown(md))
```
